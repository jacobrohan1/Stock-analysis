# Moving Average of Various Stocks

```python
google['Open'].plot(figsize=(17,5))
google['Moving Average 50'] = google['Open'].rolling(50).mean()
plt.ylabel('Opening Price')
plt.title('Moving Average for Google')
google['Moving Average 50'].plot(figsize=(17,5))
plt.legend()
```

```python
amazon['Open'].plot(figsize=(17,5))
amazon['Moving Average 50'] = amazon['Open'].rolling(50).mean()
plt.ylabel('Opening prices')
plt.title('Moving Average for Amazon')
amazon['Moving Average 50'].plot()
plt.legend()
```

```python
apple['Open'].plot(figsize=(17,5))
apple['Moving Average 50'] = apple['Open'].rolling(50).mean()
plt.ylabel('Opening Prices')
plt.title('Moving Average for Apple')
apple['Moving Average 50'].plot()
plt.legend()
```

```python
micro['Open'].plot(figsize=(17,5))
micro['Moving Average 50'] = micro['Open'].rolling(50).mean()
micro['Moving Average 100'] = micro['Open'].rolling(100).mean()
micro['Moving Average 100'].plot()
micro['Moving Average 50'].plot()
plt.ylabel('Opening Prices')
plt.title('Moving Average of 50 for Microsoft')
plt.legend()
```

# Daily Return


```python
google['Trend'] = google['Close'] - google['Open']
google['Trend'].plot(figsize = (17,5))
google['Moving Average of Daily Return'] = google['Trend'].rolling(50).mean()
google['Moving Average of Daily Return'].plot()
plt.legend()
plt.title('Daily Return for Google')
plt.ylabel('Daily Return')
```

```python
apple['Trend'] = apple['Close'] - apple['Open']
apple['Trend'].plot(figsize = (17,5))
apple['Moving Average of Daily Return'] = apple['Trend'].rolling(50).mean()
apple['Moving Average of Daily Return'].plot()
plt.ylabel('Daily Returns')
plt.title('Daily Returns of Apple')
plt.legend()
```

```python
micro['Trend'] = micro['Close'] - micro['Open']
micro['Trend'].plot(figsize = (17,5))
micro['Moving Average of Daily Return'] = micro['Trend'].rolling(50).mean()
micro['Moving Average of Daily Return'].plot()
plt.ylabel('Daily Returns')
plt.title('Daily Returns of Microsoft')
plt.legend()
```

```python
amazon['Trend'] = amazon['Close'] - amazon['Open']
amazon['Trend'].plot(figsize = (17,5))
amazon['Moving Average of Daily Return of 50'] = amazon['Trend'].rolling(50).mean()
amazon['Moving Average of Daily Return of 50'].plot()
plt.ylabel('Daily Returns')
plt.title('Daily Returns of Amazon')
plt.legend()
```
